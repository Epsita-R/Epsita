```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/HousingData.csv')
data.head()
```

```python
# check for null values
data.isnull().sum()
```

```python
# Fill null values with "NA")
data_filled = data.fillna("NA")
print(data_filled.isnull().sum())
```

```python
# Summary statistics
summary_stats = data.describe()
print(summary_stats)
```

```python
# Calculate the IQR for each column
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
# Define a function to filter outliers
def remove_outliers(df):
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
# Remove outliers
data_no_outliers = remove_outliers(data)
# Display the shape before and after removing outliers
print("Shape before removing outliers:", data.shape)
print("Shape after removing outliers:", data_no_outliers.shape)
```

```python
import matplotlib.pyplot as plt

# Scatter plot: RM vs. MEDV
plt.figure(figsize=(10, 6))
plt.scatter(data['RM'], data['MEDV'], alpha=0.7)
plt.xlabel('Average Number of Rooms per Dwelling (RM)')
plt.ylabel('Median Home Value (MEDV)')
plt.title('Relationship between RM and MEDV')
plt.grid(True)
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.impute import SimpleImputer

# Handle missing values by imputation (fill with mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(data.drop('MEDV', axis=1))
X = pd.DataFrame(X_imputed, columns=data.drop('MEDV', axis=1).columns)
y = data['MEDV']

# Correlation Analysis
correlations = data.corr()
correlations_with_target = correlations['MEDV'].abs().sort_values(ascending=False)
relevant_features_corr = correlations_with_target.index[1:]  # Excluding target variable

# Feature Importance using coefficient magnitude
model = LinearRegression()
model.fit(X, y)
coefficients = pd.Series(model.coef_, index=X.columns)
relevant_features_coeff = coefficients.abs().sort_values(ascending=False).index

# Print selected features using various methods
print("Features selected based on correlation analysis:")
print(relevant_features_corr)

print("\nFeatures selected based on coefficient magnitude:")
print(relevant_features_coeff)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into features (X) and target variable (y)
X = data.drop('MEDV', axis=1)
y = data['MEDV']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the training and testing sets
print("Training set shape - Features:", X_train.shape, "Target:", y_train.shape)
print("Testing set shape - Features:", X_test.shape, "Target:", y_test.shape)
```

```python
# Fill missing values in 'LSTAT' with mean
data['LSTAT'].fillna(data['LSTAT'].mean(), inplace=True)

# Fill missing values in other columns with mean
data.fillna(data.mean(), inplace=True)

# Fit the Linear Regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Visualize the regression line
plt.scatter(X_test['RM'], y_test, color='blue', label='Actual')
plt.plot(X_test['RM'], lr.predict(X_test), color='red', label='Predicted')
plt.xlabel('Number of Rooms (RM)')
plt.ylabel('Median Home Value (MEDV)')
plt.legend()
plt.show()
```

```python
# Model Evaluation and Interpretation
# Mean Squared Error (MSE)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Coefficient Interpretation
coefficients = pd.Series(lr.coef_, index=X.columns)
print('Coefficient for RM:', coefficients['RM'])

# Feature Transformation (square root of CRIM)
data['CRIM_sqrt'] = np.sqrt(data['CRIM'])

# Feature Interaction (RM * PTRATIO)
data['RM_PTRATIO_interaction'] = data['RM'] * data['PTRATIO']

# Model Interpretation and Real Estate Insights

# Coefficients obtained after fitting the model
feature_significance = pd.DataFrame({'Feature': X.columns, 'Coefficient': lr.coef_})
feature_significance.sort_values(by='Coefficient', ascending=False, inplace=True)
print('Feature Significance:\n', feature_significance)
```
